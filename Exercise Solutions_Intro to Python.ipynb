```python
# Exercise III: print the larger one between two numbers
m = 5
n = -99
```

```python
# Solution 1 (staighforward version)
if m > n:
  print(m)
else:
  print(n)
```

```python
# Solution 2 (if you want to print out the whole statement)
# Since m and n are integers, you have to convert them into string using str()
# Notice that the same data type (string) are concatenated with "+" under the print function
if m > n:
  print("The greater number is m = " + str(m))
else:
  print("The greater number is n = " + str(n))
```
